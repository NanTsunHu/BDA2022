## 陽明

```python
tw2609 = df[df['證券代碼'].str.contains('2609')]
tw2609['年月日'] = pd.to_datetime(tw2609['年月日']).dt.date
tw2609 = tw2609[['證券代碼', '年月日', '收盤價(元)']]
```

```python
tmp = df2020[df2020['證券代碼'].str.contains('2609')]
tmp['年月日'] = pd.to_datetime(tmp['年月日']).dt.date
tmp = tmp[['證券代碼', '年月日', '收盤價(元)']]
tw2609 = tw2609.append(tmp)
```

```python
tw2609
```

```python
tw2609 = tw2609.sort_values(by ='年月日').reset_index(drop = True)
tw2609['單日漲跌幅(%)'] = tw2609['收盤價(元)'].rolling(window=2).apply(lambda x: (x.iloc[1]-x.iloc[0])/x.iloc[0])
tw2609['label'] = tw2609['單日漲跌幅(%)'].apply(lambda x: (x>0 and '漲') or '跌')
```

```python
tw2609['label'].value_counts()
```

```python
tw2609.to_csv('C:\\Users\\user\\Desktop\\大數據與商業分析\\期中報告\\tw2609_2019-2021.csv')
```
